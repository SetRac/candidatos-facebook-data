```python
import json
from collections import Counter
import sys
import os
from dateutil import parser

#candidates = ['JohnnyArayaMonge', 'ottoguevaraguth', 'luisguillermosolisr', 'villaltaJM', 'ElDoctor2014','RodolfoPizaR']
#candidate_total_likes = dict(zip(candidates, (50301, 24146, 5334, 27262, 37495)))
candidates = ['JohnnyArayaMonge', 'ottoguevaraguth', 'luisguillermosolisr', 'villaltaJM', 'RodolfoPizaR']
candidate_total_likes = dict(zip(candidates, (90423, 39656, 37616, 93950, 19061)))

candidates_data = {}

for candidate in candidates:    
    #content = json.loads(open(os.path.join("IImed/200 last", candidate + ".last-100-fb-posts.json")).read())
    content = json.loads(open(os.path.join("IImed", candidate + ".last-500-fb-posts.json")).read())
    
    types = []
    num_posts_prior = 0
    
    for c in content['posts']['data']:
        
        dt = parser.parse(c['created_time'])

        if (dt.year, dt.month) < (2013,1): 
            num_posts_prior += 1
            continue
        
        if c.has_key('shares'):
            shares = c['shares']['count']
        else:
            shares = 0
            
        if c.has_key('likes'):
            likes = c['likes']['summary']['total_count']
        else:
            likes = 0
            
        types.append( (c['type'], c['created_time'], likes, shares, ) )

    total_likes = sum([t[2] for t in types])
    total_shares = sum([t[3] for t in types])
    
    print candidate
    print "Total Likes", candidate_total_likes[candidate]
    print "Num Posts since Jan 1, 2013 (of 100 possible)", len(types)
    print "Post Types", Counter([t[0] for t in types]).most_common(len(set([t[0] for t in types])))
    print "Total Post Likes", total_likes
    print "Total Post Shares", total_shares
    print "Oldest Post in Batch", types[-1][1]
    print "Num posts prior to Jan 1, 2013", num_posts_prior
    print "Avg likes/post", 1.0*total_likes/len(types), "({0}%)".format(1.0*total_likes/len(types)/candidate_total_likes[candidate]*100)
    print "Avg shares/post", 1.0*total_shares/len(types), "({0}%)".format(1.0*total_shares/len(types)/candidate_total_likes[candidate]*100)
    print
    
    candidates_data[candidate] = types
   
    
    
    sys.stdout.flush()
```
